```python
df = df.dropna(subset=['name', 'description'])
```

```python
df.isnull().sum()
```

```python
#5. Check Duplicate Records
df.duplicated().sum()
```

```python
#6. Standardize Column Names
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
#7. Convert Data Types
df.dtypes
```

```python
df['scraped_at'] = pd.to_datetime(df['scraped_at'])
```

```python
df['sales_volume'] = pd.to_numeric(df['sales_volume'], errors='coerce')
```

```python
df.dtypes
```

```python
#8. Check Unique Values
df['promotion'].unique()
```

```python
df['seasonal'].unique()
```

```python
df['product_position'].unique()
```

```python
df['terms'].unique()
```

```python
#9. Handle Missing Values
df[df.isnull().any(axis=1)]
```

```python
#10. Check Outliers
df['sales_volume'].describe()
```

```python
Q1 = df['sales_volume'].quantile(0.25)
Q3 = df['sales_volume'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = df[(df['sales_volume'] < lower) | (df['sales_volume'] > upper)]

outliers
```

```python
#11. Create New Features
df['promotion_flag'] = df['promotion'].map({
    'Yes':1,
    'No':0
})
```

```python
df['seasonal_flag'] = df['seasonal'].map({
    'Yes':1,
    'No':0
})
```

```python
#12. Create Sales Categories
df['sales_category'] = pd.cut(
    df['sales_volume'],
    bins=[0,1000,2000,3000,5000],
    labels=['Low','Medium','High','Very High']
)
```

```python
#13. Check Final Dataset
df.info()
```

```python
#14. Export Cleaned Dataset
df.to_csv('zara_sales_cleaned.csv', index=False)
```

Exploratory Data Analysis (EDA)****

```python
#1. Summary Statistics
df.describe()
```

```python
#2. Sales Volume Distribution
df['sales_volume'].describe()
```

```python
#3. Top Selling Products
df[['name','sales_volume']].sort_values(
    by='sales_volume',
    ascending=False
).head(10)
```

#4. Promotion Impact
Questions:

Do promoted products sell more?
What is the average sales volume for promoted vs non-promoted products?

```python

df.groupby('promotion')['sales_volume'].agg(
    ['count','mean','sum']
)
```

#5. Product Position Performance
Questions:

Does Front Store perform better than Aisle?
Which position generates the most sales?

```python

df.groupby('product_position')['sales_volume'].agg(
    ['count','mean','sum']
).sort_values(by='sum', ascending=False)
```

#6. Seasonal vs Non-Seasonal
Questions:

Do seasonal products sell better?
What percentage of sales comes from seasonal products?

```python

df.groupby('seasonal')['sales_volume'].agg(
    ['count','mean','sum']
)
```

#7. Section Analysis

```python

df.groupby('section')['sales_volume'].agg(
    ['count','mean','sum']
)
```

#8. Product Categories
Questions:

Which product categories generate the highest sales?
Jackets vs Shirts vs Pants, etc

```python

df.groupby('terms')['sales_volume'].sum().sort_values(
    ascending=False)
```
